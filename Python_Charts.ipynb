```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 2 * np.pi, 100)
y1 = x**2
y2 = np.sin(x)

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot the first function (y = x^2)
ax1.plot(x, y1, label="y = x^2")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("y = x^2")
ax1.grid(True)

# Plot the second function (y = sin(x))
ax2.plot(x, y2, label="y =sin(x)", color="orange")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("y = sin(x)")
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy import stats

# Given Dataset
data = [5, 7, 3, 8, 9, 7, 6, 7, 2, 10]

# Calculate Mean
mean = np.mean(data)

# Calculate mode
mode = stats.mode(data)

# Calculate Median
median = np.median(data)

# Calculate Variance
variance = np.var(data, ddof=0)  # Population Variance

# Calculate Standard Deviation
std_dev = np.std(data, ddof=0)  # Population standard deviation

# Print the results
print(f"Mean: {mean}")
print(f"Mode: {mode}")
print(f"Median: {median}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
x= [1,2,3,4,5,6] # time or event sequence
y= [3.2,4.8,5.1,6.3,7.0,5.8] # amplitude of earthquakes
bubble_size= [50,100,150,300,400,200] # size of bubbles

# normalize the y-values for color mapping
norm= plt.Normalize(min(y), max(y))
colors= plt.cm.viridis(norm(y))

# plot the bubbles with different colors based on the amplitude
plt.scatter(x,y,s=bubble_size,c=colors,cmap='viridis',alpha=0.7,edgecolor='k')
scatter= plt.scatter(x,y,s=bubble_size,c=colors,cmap='viridis',alpha=0.7,edgecolor='k')

# add a colorbar to show the amplitude gradient
cbar=plt.colorbar(scatter)
cbar.set_label('Amplitude')

# add labels and title
plt.xlabel('Event sequence')
plt.ylabel('Amplitude')
plt.title('Amplitude of earthquakes with bubble sizes and colors')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for demonstration
data_year = {
    'year': ['2022', '2023'],
    'revenue': [200000000, 250000000],
    'operating_profit': [-50000000, -30000000]
}

data_segment = {
    'segment': ['DS', 'DX', 'Harman', 'SDC'],
    'revenue': [50000000, 300000000, 70000000, 40000000],
    'operating_profit': [-10000000, 50000000, 2000000, 3000000]
}

# Creating dataframes
df_year = pd.DataFrame(data_year)
df_segment = pd.DataFrame(data_segment)

# Plot for revenue and operating profit by year
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot for year
df_year.plot(
    x='year',
    y=['revenue', 'operating_profit'],
    kind='bar',
    ax=ax[0]
)
ax[0].set_title('Revenue and Operating Profit by Year')
ax[0].set_ylabel('amount')

# Plot for segment
df_segment.plot(
    x='segment',
    y=['revenue', 'operating_profit'],
    kind='bar',
    ax=ax[1]
)
ax[1].set_title('Revenue and Operating Profit by Segment')
ax[1].set_ylabel('amount')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import altair as alt
import pandas as pd

# Load the data into dictionaries
data_2023 = {
    'revenue': [169992337, 66594471, 30975373, 14388454],
    'operating_profit': [14384705, -14879458, 5566478, 1173702],
    'segment': ['DX', 'DS', 'SDC', 'Harman']
}

data_2022 = {
    'revenue': [182489720, 98455270, 34382619, 13213694],
    'operating_profit': [12746074, 23815810, 5952973, 880548],
    'segment': ['DX', 'DS', 'SDC', 'Harman']
}

# Convert dictionaries into DataFrames
df_2023 = pd.DataFrame(data_2023)
df_2022 = pd.DataFrame(data_2022)

# Calculate YoY changes
df_2023['revenue_yoy'] = (df_2023['revenue'] - df_2022['revenue']) / df_2022['revenue'] * 100
df_2023['operating_profit_yoy'] = (df_2023['operating_profit'] - df_2022['operating_profit']) / df_2022['operating_profit'] * 100

# Calculate segment shares
df_2023['revenue_share'] = df_2023['revenue'] / df_2023['revenue'].sum() * 100
df_2023['operating_profit_share'] = df_2023['operating_profit'] / df_2023['operating_profit'].sum() * 100

# Melt the DataFrame for easier plotting
df_2023_melted = pd.melt(
    df_2023,
    id_vars='segment',
    value_vars=['revenue', 'operating_profit'],
    var_name='amount_type',
    value_name='amount'
)

df_2022['amount_type'] = 'revenue'
df_2022['amount'] = df_2022['revenue']
df_2022_melted = pd.concat([df_2022, df_2022.assign(amount_type='operating_profit', amount=df_2022['operating_profit'])])

# Bar chart for revenue and operating profit by year
chart1 = alt.Chart(
    pd.concat([df_2023_melted.assign(year=2023), df_2022_melted.assign(year=2022)]),
    title="Revenue and Operating Profit by Year"
).mark_bar().encode(
    x='year:O',
    y='amount',
    color='amount_type',
    tooltip=['year', 'amount_type', 'amount']
).interactive()

# Bar chart for revenue and operating profit by segment
chart2 = alt.Chart(
    pd.concat([df_2023_melted.assign(year=2023), df_2022_melted.assign(year=2022)]),
    title="Revenue and Operating Profit by Segment"
).mark_bar().encode(
    x='segment:O',
    y='amount',
    color='amount_type',
    tooltip=['segment', 'amount_type', 'amount']
).interactive()

# Bar chart for YoY changes
chart3 = alt.Chart(
    df_2023,
    title="YoY Changes in Revenue and Operating Profit by Segment"
).mark_bar().encode(
    x='segment:O',
    y=alt.Y('revenue_yoy', title='YoY Change (%)'),
    tooltip=['segment', 'revenue_yoy']
).interactive()

# Pie chart for revenue share
df_2023['percentage'] = df_2023['revenue_share'] / 100
base = alt.Chart(df_2023).encode(theta=alt.Theta('revenue_share:Q', stack=True)).properties(title='Revenue Share by Segment')

pie = base.mark_arc(outerRadius=120).encode(
    color=alt.Color('segment:N'),
    order=alt.Order('percentage', sort='descending'),
    tooltip=['segment', alt.Tooltip('revenue_share', format='.1%')],
)

text = base.mark_text(radius=140).encode(
    text=alt.Text('percentage', format='.1%'),
    order=alt.Order('percentage', sort='descending'),
    color=alt.value('black')
)

chart4 = pie + text

# Pie chart for operating profit share
df_2023['percentage'] = df_2023['operating_profit_share'] / 100
base = alt.Chart(df_2023).encode(theta=alt.Theta('operating_profit_share:Q', stack=True)).properties(title='Operating Profit Share by Segment')

pie = base.mark_arc(outerRadius=120).encode(
    color=alt.Color('segment:N'),
    order=alt.Order('percentage', sort='descending'),
    tooltip=['segment', alt.Tooltip('operating_profit_share', format='.1%')],
)

text = base.mark_text(radius=140).encode(
    text=alt.Text('percentage', format='.1%'),
    order=alt.Order('percentage', sort='descending'),
    color=alt.value('black')
)

chart5 = pie + text

# Display the charts
chart1.display()
chart2.display()
chart3.display()
chart4.display()
chart5.display()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Data
hours_studied = [3, 6, 2, 8, 4, 6, 5, 4]

# Calculating statistics
mean_value = np.mean(hours_studied)
median_value = np.median(hours_studied)
mode_value = stats.mode(hours_studied)[0]
variance_value = np.var(hours_studied, ddof=0)  # Population variance
std_dev_value = np.std(hours_studied, ddof=0)  # Population standard deviation

# Printing results
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")
print(f"Variance: {variance_value}")
print(f"Standard Deviation: {std_dev_value}")

# Bar Graph
students = range(1, len(hours_studied) + 1)
plt.figure(figsize=(7, 5))
plt.bar(students, hours_studied, color='orange')
plt.xlabel("Students")
plt.ylabel("Hours Studied")
plt.title("Study Hours of Students")
plt.xticks(students)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set theme
sns.set_theme(style="whitegrid")

# 1. Timeline of Product Evolution (Line Graph)
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2025]
models = ["OnePlus One", "OnePlus 2", "OnePlus 3", "OnePlus 5", "OnePlus 6", "OnePlus 7", "OnePlus 8", "OnePlus 9", "OnePlus 10", "OnePlus 11", "OnePlus 13"]

plt.figure(figsize=(10, 5))
plt.plot(years, np.arange(len(years)), marker='o', linestyle='-', color='red', label="OnePlus Models")
plt.xticks(years, rotation=45)
plt.yticks(np.arange(len(models)), models)
plt.title("OnePlus Product Evolution Over the Years")
plt.xlabel("Year")
plt.ylabel("Model")
plt.legend()
plt.show()


# 2. Revenue Growth Over Time (Line Graph)
revenue = [50, 150, 300, 800, 1500, 2200, 3000, 4000, 4500, 4800, 5000]  # In million USD

plt.figure(figsize=(10, 5))
sns.lineplot(x=years, y=revenue, marker="o", color='blue', label="Annual Revenue (in million $)")
plt.xlabel("Year")
plt.ylabel("Revenue ($M)")
plt.title("OnePlus Revenue Growth (2014-2025)")
plt.legend()
plt.show()


# 3. Price Comparison: OnePlus vs Competitors (Bar Chart)
models = ["OnePlus 6", "OnePlus 9", "OnePlus 11", "OnePlus 13"]
oneplus_prices = [529, 729, 799, 899]
apple_prices = [999, 1099, 1199, 1299]
samsung_prices = [949, 1049, 1149, 1249]

x = np.arange(len(models))

plt.figure(figsize=(10, 5))
plt.bar(x - 0.2, oneplus_prices, width=0.2, label="OnePlus", color='orangered')
plt.bar(x, apple_prices, width=0.2, label="Apple", color='deepskyblue')
plt.bar(x + 0.2, samsung_prices, width=0.2, label="Samsung", color='limegreen')
plt.xticks(x, models)
plt.ylabel("Price (USD)")
plt.title("Price Comparison of OnePlus vs Apple & Samsung")
plt.legend()
plt.show()


# 4. Global Market Penetration (Pie Chart)
regions = ["Asia", "Europe", "North America", "Others"]
market_share = [50, 30, 15, 5]

plt.figure(figsize=(6, 6))
plt.pie(market_share, labels=regions, autopct='%1.1f%%', colors=['lightcoral', 'deepskyblue', 'turquoise', 'gray'], startangle=140)
plt.title("OnePlus Global Market Share Distribution")
plt.show()


# 5. Value Proposition: Price vs Performance (Scatter Plot)
prices = [500, 600, 700, 800, 900, 1000, 1100]  # Price in USD
performance = [80, 85, 88, 90, 93, 96, 99]  # Arbitrary performance score

plt.figure(figsize=(8, 5))
sns.scatterplot(x=prices, y=performance, s=100, color='purple')
plt.xlabel("Price (USD)")
plt.ylabel("Performance Score")
plt.title("Price vs Performance: OnePlus vs Competitors")
plt.show()


# 6. Product Mix Breakdown (Pie Chart)
categories = ["Smartphones", "Wearables", "Accessories"]
distribution = [80, 10, 10]

plt.figure(figsize=(6, 6))
plt.pie(distribution, labels=categories, autopct='%1.1f%%', colors=['gold', 'silver', 'brown'], startangle=140)
plt.title("OnePlus Product Mix Distribution")
plt.show()
```
